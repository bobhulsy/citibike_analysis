```python
#Reading in and cleaning the data
df = pd.read_csv("Resources/2023-citibike-all.csv")
df.head()
```

```python
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])

df['trip_duration'] = df['ended_at'] - df['started_at']

df.head()
```

```python
df['trip_duration_minutes'] = df['trip_duration'].dt.total_seconds() / 60

df.head()
```

```python
#Drop columns where trip duration is less than 1 minute and start_station_id is equal to end_station_id
df = df.drop(df[(df['trip_duration_minutes'] < 1) & (df['start_station_id'] == df['end_station_id'])].index)

#filter out trips that are longer than 24 hours
df = df.drop(df[df['trip_duration_minutes'] > 1440].index)

df.to_csv("Resources/2023-citibike-all.csv", index=False)

df.head()
```

```python
#calculate the distance between the start and end stations
import numpy as np

def haversine(lat1, lon1, lat2, lon2):
    # Radius of the earth in kilometers
    R = 6371.0

    lat1_rad = np.radians(lat1)
    lon1_rad = np.radians(lon1)
    lat2_rad = np.radians(lat2)
    lon2_rad = np.radians(lon2)

    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad

    a = np.sin(dlat / 2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    return R * c

# Assuming df is your DataFrame and it has columns 'start_lat', 'start_long', 'end_lat', 'end_long'
df['distance'] = haversine(df['start_lat'], df['start_lng'], df['end_lat'], df['end_lng'])
```

```python
#Convert Distance to miles
# Convert kilometers to miles
df['distance_miles'] = df['distance'] * 0.621371

df.to_csv("Resources/2023-citibike-all.csv", index=False)

df.head()
```
